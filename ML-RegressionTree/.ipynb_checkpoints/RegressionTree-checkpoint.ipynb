## (80 pts) Task 2: Regression Tree


**(15 pts)** Let us now build the regression tree. We start by implementing the splitting criteria which is used to decide the most discriminative features at each step. We stick to lecture notes and use $RSS$ which is computed as follows:

$$RSS =\sum_{\text {left }}\left(y_{i}-y_{L}^{*}\right)^{2}+\sum_{\text {right }}\left(y_{i}-y_{R}^{*}\right)^{2}$$

where $y_L^* and y_L^*$ are mean y-value of left and right nodes.

When you implement $RSS$, pick the most correlated and least correlated feature with the output prices according to previous step. Note that correlation can be both positive and negative! For both features, compute $RSS$ for every possible split threshold and plot thresholds versus RSS scores.

Do two features display different characteristics?

```python
labelsCorrDict = dict(zip(labels[:-1], corcoef_matrix[-1:,:-1][-1]))
sortedFeatures=sorted(labelsCorrDict, key=labelsCorrDict.get, reverse=True)
leastCorr=sortedFeatures[-1]
mostCorr=sortedFeatures[0]
```

```python
def findOptSplit(rssArray):
    splitValue,splitRSS,splitIndex=rssArray[0][0],rssArray[0][1],rssArray[0][3]
    return float(splitRSS),float(splitValue),int(splitIndex)
def getRSSForSplits(feature,data):
    featureColumnIndex = int(np.where(labels == feature)[0])
    #print(data.shape)
    dataToSplit = data[np.argsort(data[:, featureColumnIndex])]
    #print(dataToSplit.shape)
    lenght=data.shape[0]
    rssForSplits=[]
    for splitIndex in range(dataToSplit.shape[0]):   
        left=dataToSplit[:splitIndex,:]   #smaller to left
        right=dataToSplit[splitIndex:,:]  #greater than or equal to right
        splitValue=dataToSplit[splitIndex,featureColumnIndex]
        splitRSS=computeRSS(left)+computeRSS(right)
        rssForSplits.append([float(splitValue),float(splitRSS),feature,splitIndex])
    rssForSplits= sorted(rssForSplits,key=itemgetter(1)) #sort accor. RSS
    optSplitRSS,optSplitValue,splitIndex=findOptSplit(rssForSplits)
    left,right=dataToSplit[:splitIndex,:],dataToSplit[splitIndex:,:]
    return left,right,optSplitRSS,optSplitValue,rssForSplits

def computeRSS(data):
    length=data.shape[1]
    mean=float(np.sum(data[:,-1]))/length
    rss=0
    for i in range(data.shape[0]):
        rss+=(data[i,-1]-mean)**2
    return rss


rssArrayLeast=getRSSForSplits(leastCorr,var_columns)[4]
rssArrayMost=getRSSForSplits(mostCorr,var_columns)[4]
rssArrayLeast,rssArrayMost=np.array(rssArrayLeast),np.array(rssArrayMost)
x1,y1=rssArrayMost[:,0], rssArrayMost[:,1]
x2,y2=rssArrayLeast[:,0], rssArrayLeast[:,1]
x1,y1 = x1.astype(np.float),y1.astype(np.float)
x2,y2 = x2.astype(np.float),y2.astype(np.float)
plt.figure()
plt.subplot(211)
plt.ylabel('RSS (most:'+str(mostCorr)+")")
plt.xlabel('Split Values')
plt.plot(x1,y1)
plt.subplot(212)
plt.ylabel('RSS(least:'+str(leastCorr)+")")
plt.xlabel('Split Values')
plt.plot(x2,y2)

plt.show()
```

**(45 pts)** Now implement the training procedure of a regression tree using $RSS$ as split criteria. Build a rule tree recursively by traversing all features and considering each split threshold to find the optimum split, at every node.

You are free to implement training procedure as a standalone function or part of a class, but in any case use maximum depth as the stopping condition.

```python

    
def buildTree(maxDepth,data,depth,featureSet,route,minLeafDataCount):
    #print(data.shape)
    if depth==maxDepth or data.shape[0]<=minLeafDataCount:
        #print(data)
        predictionDict[route]=float(np.sum(data[:,-1]))/data.shape[0] #store prediction value for this route
        return []
    minRss=math.inf
    leftData=np.array([])
    rightData=np.array([])
    splitRule=0
    for feature in featureSet:
        left,right,splitRSS,splitValue,rssArray=getRSSForSplits(feature,data)
        if splitRSS<minRss and left.shape[0]!=0 and right.shape[0]!=0:
            minRss=splitRSS
            leftData,rightData,splitRule=left,right,[feature,depth,splitValue,route,left.shape[0],right.shape[0]]
    
    #print(splitRule)
    if leftData.shape[0]==0 or rightData.shape[0]==0:
        predictionDict[route]=float(np.sum(data[:,-1]))/data.shape[0]
        return []
    """else:
        l_rules=buildTree(maxDepth,leftData,depth+1,featureSet,route+"0",minLeafDataCount)
    if rightData.shape[0]==0:
        r_rules=[]
    else:
        r_rules=buildTree(maxDepth,rightData,depth+1,featureSet,route+"1",minLeafDataCount)"""
    l_rules=buildTree(maxDepth,leftData,depth+1,featureSet,route+"0",minLeafDataCount)
    r_rules=buildTree(maxDepth,rightData,depth+1,featureSet,route+"1",minLeafDataCount)
    rules=l_rules+r_rules+[splitRule]
    #print(rules)
    return sorted(rules,key=itemgetter(3))


predictionDict={}  #gives prediction according to route at tree
rules=buildTree(maxDepth=6,data=var_columns,depth=0,featureSet=sortedFeatures,route="r",minLeafDataCount=0)  
```

**(15 pts)** Having implemented the regression tree, now write a procedure to predict test features. Given a sample, this procedure should follow the rules learned during the training to arrive at a leaf and predict the output as the mean output of the arrived leaf samples. 

If you have implemented a regression tree class, you can insert this procedure as a class function 

```python
def applyRule(rule,sample):
    feature,value=rule[0],rule[2]
    featureColumnIndex = int(np.where(labels == feature)[0])
    if sample[featureColumnIndex]<value:
        return "0"
    else:
        return "1"

def predict(rules,sample,defaultValue):  
    ruleDict = dict(zip(np.array(rules)[:,3], rules))
    route="r"
    rule=ruleDict.get(route)
    while not (rule is None):
        route+=applyRule(rule,sample)
        rule=ruleDict.get(route)
    prediction=predictionDict.get(route)
    if prediction is None:  #in ideal case there won't be any None prediction! 
        print("ROUTE :"+route)
        print("SAMPLE :")
        print(sample)
        print(rules)
        prediction=defaultValue
    return prediction

def testDataChecker(rules,data,defaultValue):   #gets rules and data then returns 
    lenght=data.shape[0]
    predictions=[0 for i in range(lenght)]
    for i in range(lenght):
        sample=data[i,:]
        predictions[i]=predict(rules,sample,defaultValue)
    results=np.concatenate(([data[:,-1]],[np.array(predictions)]), axis=0)  #predictions along with real data
    return results   #index 0:real value index 1: prediction
```

```python

results=testDataChecker(rules,test_data,defaultValue=-1)
realData=results[0] #results for test_data 
predictions=results[1]  #predictions for test_data at corresponding index
```

## **(10 pts)** Task 3: Evaluation

**(5 pts)** Now is time to pick the best maximum depth and observe your tree's performance! Implement a 5-fold cross validation procedure to experiment with maximum depths from 3 to 10. Report mean and standard deviation for each depth and pick the best one. For comparison you can use $R^2$, which is a metric frequently used to evaluate regression models. You can use `r2_score` function of `scikit-learn` and read more [here](https://scikit-learn.org/stable/modules/model_evaluation.html#r2-score).

```python
from sklearn.metrics import r2_score
def k_fold_cross_validation(k, X,minDepth,maxDepth,defaultValueforPredictions):
    
    N = X.shape[0]
    part_N = int(np.floor(N / k))
    
  
    values = []  # this list will hold the mean,std dev,r2_score in each validation
    
    
    for i in range(k):
        
        # use i'th part as the validation set
        
        validation_X = X[i*part_N:(i+1)*part_N,:]
        training_X = np.concatenate((X[:i*part_N,:],X[(i+1)*part_N:,:]), axis=0) 
        #print(validation_X.shape, training_X.shape)
        for depth in range(minDepth,maxDepth+1):
            predictionDict={}  #gives prediction according to route at tree
            rules=buildTree(maxDepth=depth,data=training_X,depth=0,featureSet=sortedFeatures,route="r",minLeafDataCount=0)
            results=testDataChecker(rules=rules,data=validation_X,defaultValue=defaultValueforPredictions)
            realData=results[0]
            predictions=results[1]
            std,mean=np.std(realData, axis=0),np.mean(realData, axis=0)
            values.append(["ValidationSet "+str(i),"Depth: "+str(depth),"r2 Score: ",r2_score(realData,predictions),"Std: "+str(std),"Mean: "+str(mean)])
    
        
    return sorted(values,key=itemgetter(3),reverse=True)





```

```python
# This part gonna take some time. 
values=k_fold_cross_validation(5,X_train,minDepth=3,maxDepth=10,defaultValueforPredictions=-1)
#There won't be fail in prediction in our case, defaultValueforPredictions is just for generalization
#You can change it's value, there won't be any effect.
print("Results for cross-validation sorted by r2_score: ")
print("Best r2_score combination : "+ str(values[0]))
print(values)
```
